```python
import numpy as np
from geneticalgorithm import geneticalgorithm as ga
```

```python
def f(x):
    
    if (
            not -x[0] + 2*x[1]*x[0] <= 8
            or 
            not 2*x[0] + x[1] <= 14
            or 
            not 2*x[0] - x[1] <= 10
        ):
        penalty = np.inf
        
    else:
        penalty = 0
        
    return -(x[0] + x[1] * x[0]) + penalty
```

```python
variables_bounds = np.array([[0,10], [0,10]])
variables_type = np.array([['int'], ['real']])
model = ga(function = f, dimension = 2, variable_type_mixed = variables_type, variable_boundaries = variables_bounds)
model.run()
```
